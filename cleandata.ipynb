```python
import pandas as pd

df = pd.read_csv("fr-en-annuaire-education.csv", sep=";", dtype=str)

df["Type_etablissement"] = df["Type_etablissement"].fillna("")

df_lycees = df[df["Type_etablissement"].str.contains("Lycée", case=False, na=False)]

df_lycees.to_csv("lycees_seulement.csv", index=False)

print("Nombre de lycées :", len(df_lycees))
```
